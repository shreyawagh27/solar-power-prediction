```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/solar_weather_dataset.csv")

df.head()
```

```python
df["timestamp"] = pd.to_datetime(df["timestamp"])
```

```python
df["hour"] = df["timestamp"].dt.hour

df["day_of_week"] = df["timestamp"].dt.dayofweek

df["is_weekend"] = df["day_of_week"].isin([5,6]).astype(int)

df.head()
```

```python
df["sin_hour"] = np.sin(2 * np.pi * df["hour"] / 24)

df["cos_hour"] = np.cos(2 * np.pi * df["hour"] / 24)
```

```python
df["temp_cloud_interaction"] = df["temperature"] * (1 - df["cloud_cover"]/100)
```

```python
df["rolling_temp"] = df["temperature"].rolling(3).mean()
```

```python
df["lag_solar"] = df["solar_output"].shift(1)
```

```python
df = df.fillna(0)
```

```python
df.head()
```

```python
df.to_csv("../data/processed_solar_data.csv", index=False)

print("Processed dataset saved!")
```
